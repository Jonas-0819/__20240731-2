```python
import requests
from requests.exceptions import HTTPError,ConnectionError,Timeout
try:
    url='https://raw.githubusercontent.com/roberthsu2003/__2024_07_29_1_3_night__/refs/heads/main/lesson14/names.txt'
    response=requests.get(url)
    response.raise_for_status()
    #print(type(response.text))
    content=response.text
    content
except ConnectionError:
    print("連線伺服器發生錯誤")
except Timeout:
    print("伺服器太忙,沒有回應")
except HTTPError:
    print("伺服器回應連線錯誤")
except Exception:
    print("不知名的錯誤")
file=open('names.txt',encoding='utf-8',mode='w',newline='')
file.write(content)
file.close()
```

```python
import random
with open('names.txt',encoding='utf-8') as file:
    names_str = file.read()
    display(names_str)
names_list = names_str.split(sep='\n')
names = random.choices(names_list,k=50)
```

```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
df1 = pd.DataFrame(scores,
                   columns=['國文','英文','數學','地理','探索'],
                   index=names)
df1
```

```python
#to_numpy()
df1.to_numpy()
```

```python
#to_csv()
df1.to_csv('students.csv')
```
